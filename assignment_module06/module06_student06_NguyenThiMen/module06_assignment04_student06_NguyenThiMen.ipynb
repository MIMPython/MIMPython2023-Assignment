```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
```

```python
input_image = img.imread(
    'additionalFolder/module06_assignment04_student06_NguyenThiMen_mooncake.png')
```

```python
input_image.shape
```

```python
h, w, c = input_image.shape
```

```python
w -= 5 
```

```python
img1 = input_image[:h//2 - 1, :w//2, :] # (248, 250, 4)
plt.imshow(img1);
```

```python
img2 = input_image[:h//2 -1, w//2 + 2:w+1, :] # (248, 249, 4)
plt.imshow(img2);
```

```python
img3 = input_image[h//2 + 1:h, 2:w//2 , :] # (248, 248, 4)
plt.imshow(img3);
```

```python
img4 = input_image[h//2+1:h, w//2 + 2:w+1, :] # (248, 249, 4)
plt.imshow(img4);
```

```python
img5 = np.concatenate((img2, img4), axis = 1) # (248, 498, 4)
plt.imshow(img5);
```

```python
img7 = np.concatenate((img5, img6), axis = 0)
plt.axis('off')
plt.imshow(img7);
plt.savefig('additionalFolder/module06_assignment04_student06_NguyenThiMen_result_mooncake.png', format = 'png')
```
